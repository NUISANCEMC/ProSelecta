```python
from pyHepMC3 import HepMC3 as hm3
import pyProSelecta as pps
```

```python
rdr = hm3.deduce_reader("neut.vect.hepmc")
evt = hm3.GenEvent()

i = 0
while not rdr.failed():
  rdr.read_event(evt)
  if rdr.failed():
    break

  print(f"---- Event {i}")

  prots = pps.event.all_out_part(evt,2212)
  print("protons: ", [pps.p3mod(p) for p in prots])
  prots_sorted = pps.part.sort_ascending(pps.p3mod,prots)
  print("prots_sorted: ", [pps.p3mod(p) for p in prots_sorted])
  prots_filtered = pps.part.filter(pps.p3mod < 0.5 * pps.unit.GeV, prots)
  print("prots_filtered: ", [pps.p3mod(p) for p in prots_filtered])
  pips, pims = pps.event.all_out_part(evt,[211,-211])
  print("pips: ", [pps.p3mod(p) for p in pips])
  print("pims: ", [pps.p3mod(p) for p in pims])

  print("all out: ", [f"{p.pid()}, {pps.p3mod(p)}" for p in pps.event.all_out_part(evt)])
  all_out = pps.part.sum(pps.momentum ,pps.event.all_out_part(evt))
  print("sumpt: ", all_out.pt())

  print("----")
  if i > 10:
      break
  i += 1
```
